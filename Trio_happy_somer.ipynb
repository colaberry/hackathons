```python
import pandas as pd
train = pd.read_csv("https://data.somervillema.gov/api/views/yevj-2b33/rows.csv")
train_2011 = train.where(train['Year'] == 2011)
train_2013 = train.where(train['Year'] == 2013)
train_2015 = train.where(train['Year'] == 2015)
train_2011.dropna(axis = 1, how= 'all', inplace = True)
train_2013.dropna(axis = 1, how= 'all', inplace = True)
train_2015.dropna(axis = 0, how = 'all', inplace = True)
train_2013.dropna(axis = 0, how = 'all', inplace = True)
train_2011.dropna(axis = 0, how = 'all', inplace = True)
train_2015.dropna(axis = 1, how= 'all', inplace = True)
train_2015_publicschools = train_2015['How.would.you.rate.the.following..The.overall.quality.of.public.schools.']
train_2015.drop(['How.would.you.rate.the.following..The.overall.quality.of.public.schools.'], axis =1, inplace = True)
train_2013.dropna(axis = 1, how= 'all', inplace = True)
train_2013_publicschools = train_2013['How.would.you.rate.the.following..The.overall.quality.of.public.schools.']
train_2013.drop(['How.would.you.rate.the.following..The.overall.quality.of.public.schools.'], axis =1, inplace = True)
train_2015 = train_2015.apply(lambda x:x.fillna(x.value_counts().index[0]))
train_2013 = train_2013.apply(lambda x:x.fillna(x.value_counts().index[0]))
```
