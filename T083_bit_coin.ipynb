### Lesson_5:Predicting the values

There mainly 4 steps in building a model

1.Defining a model

2.Fitting a model

3.Predicting the values

4.Evaluating(How accurate???)
#### Exercise::Add  "btc_trade_volume","btc_blocks_size","btc_hash_rate" to the cols_to_use1 and assign the value to mae1

```python
#choosing a model
from sklearn.ensemble import GradientBoostingRegressor
cols_to_use1=["btc_market_cap","btc_miners_revenue","btc_transaction_fees","btc_n_transactions",
             "btc_n_transactions_total","btc_estimated_transaction_volume_usd","btc_total_bitcoins"]
X=bitcoin_data1[cols_to_use1]
y=bitcoin_data1["btc_market_price"]
#splitting the given data into training and testing
from sklearn.model_selection import train_test_split
train_X,test_X,train_y,test_y=train_test_split(X,y,train_size=0.7,test_size=0.3,random_state=0)
my_model=GradientBoostingRegressor()
#fitting a model
my_model.fit(train_X,train_y)
#predicting the values
predictions=my_model.predict(test_X)
#evaluating the predictions
from sklearn.metrics import mean_absolute_error
print("MAE::"+str(mean_absolute_error(predictions,test_y)))
```
