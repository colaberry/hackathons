```python
import pandas as pd
import matplotlib as mp
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels as sp
import sklearn.metrics as m
dat=(pd.read_csv("https://raw.githubusercontent.com/colaberry/data/master/Bitcoin/bitcoin_dataset.csv"))
date_data=dat[dat['Date'] != '########']
d=pd.DataFrame(date_data,columns=["Date","btc_total_bitcoins"])
sns.tsplot(d['btc_total_bitcoins'], time=d.index, condition='total bitcoins in 10000000', color='red')
plt.ylabel('Total number of bitcoins')
plt.xlabel('Days')
plt.title('Total number of bitcoins in days')
```
