### Observation:
since this is a standard clustering problem we can apply both K-means and Heirarchial Clustering techniques to solve this.
            
### Here we will first find the optimal no.of clusters using the elbow method

this helps us to select optimal number of clusters for classification.
Initialize variable q with the appropriate no.of clusters by observing the graph.



```python
wcss=[]

for i in range (1,10):
    cl=KMeans(init='k-means++',n_clusters=i,n_init=10,max_iter=300)
    pred=cl.fit(train)
    wcss.append(cl.inertia_)

plt.plot(np.arange(1,10,1),wcss,color='red')

plt.xlabel("No.of Clusters")
plt.ylabel("WCSS ")
plt.title("Finding the optimal number of clsters(Elbow Method)")
plt.show()
q=3
```

```python
cl=KMeans(n_clusters=5,init='k-means++',n_init=10,max_iter=300)
y_pred=cl.fit_predict(train)
#this is used for clustering purpose

#visualization of result using the plot between salary and expenditure

plt.scatter(X[y_pred==0,[0]],X[y_pred==0,[1]],c='r',s=50,label='5th level')
plt.scatter(X[y_pred==1,[0]],X[y_pred==1,[1]],c='b',s=50,label='4 th level')
plt.scatter(X[y_pred==2,[0]],X[y_pred==2,[1]],c='y',s=50,label='3rd level')
plt.scatter(X[y_pred==3,[0]],X[y_pred==3,[1]],c='c',s=50,label='2nd level')
plt.scatter(X[y_pred==4,[0]],X[y_pred==4,[1]],c='w',s=50,label='1st level')
plt.scatter(cl.cluster_centers_[:,0],cl.cluster_centers_[:,1],s=100,label='Centroids',c='g')
plt.legend()
plt.xlabel("Income")
plt.ylabel("Expenditure")
plt.title("Formation of clusters")
plt.show()
```
