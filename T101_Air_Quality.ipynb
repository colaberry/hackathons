Import the neccesary libraries required to read the CSV file 
i.e. pandas libary 
Read the file using read_CSV method


```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/colaberry/data/master/8hour_42101_2017_10K.csv')
df
```

Import the matplotlib package to plot the graph showing the relation between the variables in the dataset file


```python
import matplotlib.pyplot as plt
df.plot(x='Observation Count',y='Mean Including All Data')
```

* Performing clustering *
import sklearn for using kmeans algorithm to implement clustering on the data
import numpy for creating N-dimensional arrays


```python

# clustering dataset
from sklearn.cluster import KMeans
from sklearn import metrics
import numpy as np
import matplotlib.pyplot as plt
f1 = df['Observation Count'].values
f2 = df['Mean Including All Data'].values

plt.plot()
plt.xlim([0,12])
plt.ylim([0,12])
plt.title('Dataset')
plt.xlabel('Observation Count')
plt.ylabel('Mean Including All Data')
plt.scatter(f1, f2)
plt.show()

plt.plot()
X = np.array(list(zip(f1, f2))).reshape(len(f1), 2)
colors = ['b', 'g', 'r']
markers = ['o', 'v', 's']

# KMeans 
K = 3
kmeans_model = KMeans(n_clusters=K).fit(X)

plt.plot()
for i, l in enumerate(kmeans_model.labels_):
    plt.plot(f1[i], f2[i], color=colors[l], marker=markers[l],ls='None')
    plt.xlim([0,12])
    plt.ylim([0,12])

plt.xlabel('Observation Count')
plt.ylabel('Mean Including All Data')
plt.show()
```
